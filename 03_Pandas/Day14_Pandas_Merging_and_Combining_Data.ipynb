# 📂 Practice Datasets

Today's concepts will be practiced using multiple employee datasets that simulate information coming from different departments or business systems.

These datasets will later be combined using different Pandas techniques.

```python
import pandas as pd

# Employee Information
employees = pd.DataFrame({
    "EmpID":[101,102,103,104],
    "Name":["Alice","Bob","Charlie","David"],
    "Department":["HR","IT","Finance","IT"]
})

# Salary Information
salary = pd.DataFrame({
    "EmpID":[101,102,103,104],
    "Salary":[45000,70000,90000,65000]
})

# Bonus Information
bonus = pd.DataFrame({
    "EmpID":[101,102,104],
    "Bonus":[5000,7000,6000]
})

# New Employees
new_employees = pd.DataFrame({
    "EmpID":[105,106],
    "Name":["Eva","Frank"],
    "Department":["HR","Finance"]
})
```

# 📖 Part 1 – Combining Data with `concat()`

**Objective:** Learn how to combine multiple datasets vertically and horizontally using `concat()`.

## ✅ Tasks

Using the Practice Datasets:

- Append new employees below the existing employee dataset.
- Concatenate datasets row-wise.
- Concatenate datasets column-wise.
- Observe how indexes behave before and after concatenation.
- Reset indexes where necessary.

```python
employees_copy = pd.concat([employees,new_employees])
print("Append new employees : \n",employees_copy)
employees_concat = pd.concat([employees,salary,bonus],axis=1)
print("\nConcatenate datsets column-wise : \n",employees_concat)
```

### 📝 Observation

- `concat()` combines multiple DataFrames either row-wise or column-wise.
- Row-wise concatenation appends records, while column-wise concatenation adds new columns based on index alignment.
- Resetting indexes after concatenation helps maintain a clean and organized dataset.

# 📖 Part 2 – Understanding `merge()`

**Objective:** Learn how related datasets are combined using a common key.

## ✅ Tasks

Using the Practice Datasets:

- Merge Employee Information with Salary Information.
- Merge Employee Information with Bonus Information.
- Specify the common key explicitly.
- Observe how matching records are combined into a single dataset.

```python
#Merge Employee Information with Salary Information
employees_merge = pd.merge(employees,salary)
print("Merge Employee Information with Salary Information : \n",employees_merge)
```

```python
#Merge Employee Information with bonus Information
employees_merge_bonus = pd.merge(employees_merge,bonus)
print("Merge Employee Information with Bonus Information : \n",employees_merge_bonus)
```

```python
#Merge Employee Information with bonus Information
employees_merge_bonus = pd.merge(employees_merge,bonus,how="left")
print("Merge Employee Information with Bonus Information : \n",employees_merge_bonus)
```

### 📝 Observation

- `merge()` combines related datasets using a common key.
- Matching records from different DataFrames are joined into a single dataset.
- Using explicit keys makes data integration accurate and easier to understand.

# 📖 Part 3 – Types of Joins

**Objective:** Understand how different join types affect the final dataset.

## ✅ Tasks

Using Employee Information and Bonus Information:

Perform:

- Inner Join
- Left Join
- Right Join
- Outer Join

Compare the output of each join and identify which records are retained or discarded.

```python
#Merge Employee Information with bonus Information inner join
employees_merge_bonus = pd.merge(employees_merge,bonus,how="inner")
print("Merge Employee Information with Bonus Information : \n",employees_merge_bonus)
```

```python
#Merge Employee Information with bonus Information with right join
employees_merge_bonus = pd.merge(employees_merge,bonus,how="right")
print("Merge Employee Information with Bonus Information : \n",employees_merge_bonus)
```

```python
#Merge Employee Information with bonus Information with outer join
employees_merge_bonus = pd.merge(employees_merge,bonus,how="outer")
print("Merge Employee Information with Bonus Information : \n",employees_merge_bonus)
```

### 📝 Observation

- Different join types determine which records are retained during the merge operation.
- Inner joins keep only matching records, while left, right, and outer joins preserve additional unmatched data based on the selected join type.
- Choosing the correct join type is essential for producing accurate analytical results.

# 🧩 Mini Challenge – Student Performance Report

## Scenario

A school stores student information in separate tables.

Create the following datasets:

### Students

- Student ID
- Name
- Class

### Marks

- Student ID
- Mathematics
- Science

### Attendance

- Student ID
- Attendance Percentage

Using Pandas:

- Merge all three datasets.
- Display one complete student report.
- Calculate the average marks.
- Identify students scoring above the class average.

Write a short summary of your findings.

```python
students = pd.DataFrame({"stuid" : [1,2,3,4,5,6],
                         "Name" : ["Anirudh","Karthik","Mani","Poorna","Rithvik","Paddu"],
                         "Class" : [4,5,6,7,8,9]})

Marks = pd.DataFrame({"stuid" : [1,2,3,4,5,6],
                         "Mathematics" : [75,93,90,70,93,85],
                         "Science" : [64,85,76,67,88,49]})

Attendance = pd.DataFrame({"stuid" : [1,2,3,4,5,6],
                         "Attendance" : [90,70,93,85,75,93]})

merge_all = pd.merge(students,Marks)
merge_all = pd.merge(merge_all,Attendance)

print(merge_all)
class_average = merge_all[["Mathematics","Science"]].mean(axis=1).mean()
print("\nAverage Marks : ",class_average)

print("\nStudents scoring above class average : \n",merge_all[merge_all[["Mathematics","Science"]].mean(axis=1)>class_average].Name)
```

## 📌 Mini Challenge Summary

I combined separate student, marks, and attendance datasets into a single report using `merge()`. After creating a unified dataset, I calculated the class average and identified students whose average marks exceeded the overall class average.

This exercise demonstrated how multiple related datasets can be integrated to perform meaningful academic performance analysis and generate actionable insights.

# 🚀 Bonus Challenge – Retail Sales Dashboard

## Scenario

A retail company stores information in separate tables.

Create four datasets:

### Products

- Product ID
- Product Name
- Category

### Sales

- Product ID
- Quantity Sold

### Prices

- Product ID
- Unit Price

### Ratings

- Product ID
- Customer Rating

Using Pandas:

- Merge all datasets into one.
- Create a Revenue column.
- Identify the highest revenue product.
- Identify the highest-rated product.
- Sort products by revenue.

Finally,

Write **five business insights** based on the combined dataset.

```python
products = pd.DataFrame({"prodid" : [1,2,3],
                         "ProdName" : ["Realme Buds","OnePlus Tv","Redmi 12 pro"],
                         "Category" : ["Wearables","TV","Mobiles"]})

Sales = pd.DataFrame({"prodid" : [1,2,3],
                         "Quantity Sold" : [75,93,90]})

Prices = pd.DataFrame({"prodid" : [1,2,3],
                         "Unit Price" : [900,70000,20000]})

Ratings = pd.DataFrame({"prodid" : [1,2,3],
                     "Customer Ratings" : [8,8.5,9.5]})

retail_merge = pd.merge(products,Sales)
retail_merge = pd.merge(retail_merge,Prices)
retail_merge = pd.merge(retail_merge,Ratings)

print(retail_merge)

retail_merge["Revenue"] = retail_merge["Quantity Sold"]*retail_merge["Unit Price"]

print("\n",retail_merge)
```

```python
#Highest revenue product
print("Highest revenue product :\n",retail_merge.iloc[retail_merge["Revenue"].argmax()])

#Highest rated product
print("\nHighest rated product : \n",retail_merge.iloc[retail_merge["Customer Ratings"].argmax()])
```

```python
#Sort by revenue
print("sorted by revenue : \n",retail_merge.sort_values(by="Revenue",ascending=False))
```

## 💡 Business Insights

1. **OnePlus TV** generated the highest revenue due to its high selling price combined with strong sales volume, making it the most valuable product.

2. **Redmi 12 Pro** received the highest customer rating, indicating excellent customer satisfaction and strong product acceptance.

3. **Realme Buds** sold a good number of units but contributed comparatively lower revenue because of its lower unit price.

4. Combining product, sales, pricing, and customer rating data provides a complete view of business performance that would not be possible from individual datasets.

5. Revenue analysis together with customer ratings enables businesses to identify products that are both profitable and well-received, supporting better pricing and inventory decisions.
