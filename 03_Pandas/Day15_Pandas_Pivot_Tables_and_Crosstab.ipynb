# 📂 Practice Dataset

Today's concepts will first be practiced using a small retail sales dataset.

The dataset contains sales transactions from different regions, product categories, sales representatives, and months. This structure will help demonstrate how pivot tables summarize numerical values across multiple business dimensions.

The Titanic dataset will be used only during the Mini Challenge and Bonus Challenge.

```python
import pandas as pd

sales_data = {
    "Region": [
        "North", "South", "East", "West",
        "North", "South", "East", "West",
        "North", "South", "East", "West"
    ],
    
    "Category": [
        "Electronics", "Electronics", "Furniture", "Furniture",
        "Furniture", "Furniture", "Electronics", "Electronics",
        "Electronics", "Furniture", "Furniture", "Electronics"
    ],
    
    "Salesperson": [
        "Alice", "Bob", "Charlie", "David",
        "Alice", "Bob", "Charlie", "David",
        "Alice", "Bob", "Charlie", "David"
    ],
    
    "Month": [
        "January", "January", "January", "January",
        "February", "February", "February", "February",
        "March", "March", "March", "March"
    ],
    
    "Units_Sold": [
        20, 18, 15, 12,
        14, 17, 22, 19,
        25, 16, 18, 21
    ],
    
    "Revenue": [
        50000, 42000, 35000, 30000,
        32000, 39000, 52000, 47000,
        61000, 37000, 41000, 54000
    ]
}

df = pd.DataFrame(sales_data)

print(df)
```

# 📖 Part 1 – Creating Basic Pivot Tables

**Objective:** Learn how to summarize numerical values across categories using `pivot_table()`.

## ✅ Tasks

Using the Practice Dataset:

- Create a pivot table showing total Revenue by Region.
- Create a pivot table showing average Revenue by Category.
- Create a pivot table showing total Units Sold by Salesperson.
- Explicitly specify the `values`, `index`, and `aggfunc` parameters.
- Compare the pivot table output with the original transaction-level dataset.

Focus on understanding how raw rows are converted into grouped summaries.

```python
pivot1 = pd.pivot_table(df,index="Region",values="Revenue",aggfunc="sum")
print("Pivot table showing total Revenue by region : \n",pivot1)

pivot2 = pd.pivot_table(df,index="Category",values="Revenue")
print("\nPivot table showing Average Revenue by Category : \n",pivot2)

pivot3 = pd.pivot_table(df,index="Salesperson",values="Units_Sold",aggfunc="sum")
print("\nPivot table showing total units sold by Salesperson : \n",pivot3)
```

### 📝 Observation

- `pivot_table()` summarizes numerical data by grouping values into meaningful business categories.
- The `values`, `index`, and `aggfunc` parameters define what is summarized, how it is grouped, and which calculation is performed.
- Pivot tables provide a cleaner and more report-oriented view than raw transaction-level data.

# 📖 Part 2 – Multi-Dimensional Pivot Tables

**Objective:** Learn how to compare one category across another by placing variables in rows and columns.

## ✅ Tasks

Using the Practice Dataset:

- Create a pivot table showing total Revenue by Region and Category.
- Place Region in the rows.
- Place Category in the columns.
- Create another pivot table showing Units Sold by Month and Category.
- Observe how adding the `columns` parameter changes the report structure.
- Identify combinations where no matching transaction exists.

Focus on understanding how pivot tables create two-dimensional business reports.

```python
pivot4 = pd.pivot_table(df,index=["Region"],columns="Category",values="Revenue",aggfunc="sum")
print("\nPivot table showing total Revenue by Region and Category : \n",pivot4)

pivot5 = pd.pivot_table(df,index=["Month"],columns="Category",values="Units_Sold",aggfunc="sum")
print("\nPivot table showing total Units sold by Month and Category : \n",pivot5)
```

### 📝 Observation

- Adding the `columns` parameter creates a two-dimensional report that compares data across multiple categories.
- Multi-dimensional pivot tables make it easier to identify relationships and trends between different business dimensions.
- Missing category combinations appear as `NaN` unless handled explicitly.

# 📖 Part 3 – Multiple Aggregations in Pivot Tables

**Objective:** Learn how to calculate multiple summary statistics within a single pivot table.

## ✅ Tasks

Using the Practice Dataset:

Create a pivot table grouped by Category that calculates:

- Total Revenue
- Average Revenue
- Maximum Revenue
- Total Units Sold
- Average Units Sold

Then create another report grouped by Region using multiple aggregation functions.

Observe how a single pivot table can provide several business metrics simultaneously.

```python
pivot6 = pd.pivot_table(df,index=["Category"],values=["Revenue","Units_Sold"],aggfunc={"Revenue":["sum","mean","max"],"Units_Sold":["sum","mean"]})
print("\nPivot table grouped by Category using multiple aggregations : \n",pivot6)

pivot7 = pd.pivot_table(df,index=["Region"],values=["Revenue","Units_Sold"],aggfunc={"Revenue":["sum","mean","max"],"Units_Sold":["sum","mean"]})
print("\nPivot table grouped by Region using multiple aggregations : \n",pivot7)
```

### 📝 Observation

- Multiple aggregation functions can be calculated simultaneously within a single pivot table.
- Different metrics such as total, average, and maximum provide a more comprehensive understanding of business performance.
- Consolidating multiple statistics into one report improves analytical efficiency and readability.

# 📖 Part 4 – Improving Pivot Table Reports

**Objective:** Learn how to make pivot tables more complete and readable using totals and missing-value controls.

## ✅ Tasks

Using a Region and Category pivot table:

- Add row and column totals using `margins=True`.
- Change the total label using `margins_name`.
- Replace missing combinations using `fill_value`.
- Round numerical results where necessary.
- Give the completed report a clear variable name.
- Compare the report before and after applying these options.

Focus on improving the readability and usefulness of pivot tables for business reporting.

```python
print(pivot4)
pivot8 = pd.pivot_table(df,index=["Region"],columns="Category",values="Revenue",aggfunc="sum",margins=True,margins_name="Grand Total",fill_value=1)
print("\nPivot table showing total Revenue by Region and Category : \n",pivot8)
```

### 📝 Observation

- Additional parameters such as `margins`, `margins_name`, and `fill_value` enhance the readability and completeness of pivot tables.
- Grand totals provide a quick overview of overall business performance.
- Replacing missing values and formatting reports makes the output more suitable for decision-making and presentation.

# 📖 Part 5 – Frequency Analysis with `crosstab()`

**Objective:** Learn how to analyze the frequency and distribution of categorical variables using `pd.crosstab()`.

## ✅ Tasks

Using the Practice Dataset:

- Create a frequency table between Region and Category.
- Create a frequency table between Month and Category.
- Add row and column totals.
- Create a normalized crosstab showing row-wise proportions.
- Compare the purpose of `crosstab()` with `pivot_table()`.
- Identify which function is more appropriate for counts and which is more appropriate for numerical summaries.

Focus on understanding category distributions rather than revenue calculations.

```python
freq_table1 = pd.crosstab(index=df["Region"],columns=df["Category"])
print("Frequency Table b/w Region and Category :\n",freq_table1)

freq_table2 = pd.crosstab(index=df["Month"],columns=df["Category"],margins=True,margins_name="Total")
print("\nFrequency Table b/w Month and Category :\n",freq_table2)


freq_table3 = pd.crosstab(index=df["Month"],columns=df["Category"],margins=True,margins_name="Total",normalize="index")
print("\nNormalised Frequency Table b/w Month and Category :\n",freq_table3)
```

### 📝 Observation

- `crosstab()` is designed to analyze the frequency distribution between categorical variables.
- Normalized crosstabs display proportions or percentages instead of raw counts, providing deeper insights into category distributions.
- While pivot tables summarize numerical values, crosstabs focus on understanding categorical relationships.

# 🚀 Bonus Challenge – Titanic Management Dashboard

## Scenario

## Report 1 – Passenger Profile

Create a pivot table containing:

- Passenger Class as rows
- Average Age
- Average Fare
- Passenger Count

Use multiple aggregation functions within one report.

---

```python
pivot_table5 = pd.pivot_table(df1,index="Pclass",values=["Age","Fare","PassengerId"],aggfunc={"Age":"mean","Fare":"mean","PassengerId":"count"})

pivot_table5 = pivot_table5.rename(columns={
    "Age": "Average Age",
    "Fare": "Average Fare",
    "PassengerId": "Total Passengers"
})

print("Passenger Profile : \n",pivot_table5)
```

## Report 2 – Gender and Class Performance

Create a pivot table showing:

- Passenger Class and Sex as grouped rows
- Average Fare
- Average Age
- Survival Rate

Remember that the mean of a binary survival column represents the proportion of passengers who survived.

---

```python
pivot_table6 = pd.pivot_table(df1,index=["Pclass","Sex"],values=["Age","Fare","Survived"],aggfunc={"Age":"mean","Fare":"mean","Survived":"mean"})

pivot_table6 = pivot_table6.rename(columns={
    "Age": "Average Age",
    "Fare": "Average Fare",
    "Survived": "Survival Rate"
})

print("Gender and Class Performance : \n",pivot_table6)
```

## Report 4 – Survival Crosstab

Create a crosstab between:

- Sex
- Survival Status

## Report 5 – High-Fare Passenger Analysis

Create a new Fare Category column using suitable boundaries such as:

- Low Fare
- Medium Fare
- High Fare

Then create a pivot table showing:

- Fare Category as rows
- Passenger Class as columns
- Average Survival Rate as values

---

```python
def fare_category(fare):
    if fare>25:
        return "High Fare"
    elif 10 <= fare <= 25:
        return "Medium Fare"
    elif fare<10:
        return "Low Fare"
df1["Fare Category"] = df1["Fare"].apply(fare_category)

pivot_table8 = pd.pivot_table(df1,index=["Fare Category"],columns="Pclass",values=["Survived"],aggfunc="mean",fill_value=0)

print("High Fare Passenger Analysis : \n",pivot_table8)
```

# 🌟 Analyst Extension Challenge – Answers

### 1. What is the average fare for each passenger class?

**Answer:** I would use `groupby()` because I am calculating a single aggregation (average fare) for each passenger class. It provides a simple and efficient grouped summary.

---

### 2. How many male and female passengers belonged to each passenger class?

**Answer:** I would use `pivot_table()` because I need to compare passenger counts across two categorical dimensions (Passenger Class and Sex) in a report-style format.

---

### 3. What percentage of passengers survived within each gender?

**Answer:** I would use `crosstab()` because it can easily calculate normalized frequencies (percentages) between categorical variables, making it ideal for distribution analysis.

---

### 4. What are the mean, maximum, and minimum fares for each embarkation port?

**Answer:** I would use `pivot_table()` because it supports multiple aggregation functions within a single report, allowing mean, maximum, and minimum fares to be calculated simultaneously.

---

### 5. How does average survival vary across fare categories and passenger classes?

**Answer:** I would use `pivot_table()` because it summarizes numerical values (average survival) across multiple categorical dimensions, making comparisons straightforward.

---

### 6. How many passengers embarked from each port?

**Answer:** I would use `groupby()` because I only need to calculate the count for a single categorical column. A simple grouped count is sufficient.

---

### 7. Which passenger class had the highest average age?

**Answer:** I would use `groupby()` because I only need the average age for each passenger class, followed by identifying the highest value. Although a pivot table could also solve this problem, `groupby()` is the simpler and more direct approach.

---

### 8. How are passengers distributed across both gender and survival status?

**Answer:** I would use `crosstab()` because it is specifically designed to analyze the frequency distribution between two categorical variables.
