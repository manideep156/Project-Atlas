# 📂 Practice Dataset

Today's concepts will first be practiced on a small dataset where every change is easy to observe and verify.

This allows me to understand the behavior of each Pandas function before applying the same concepts to a real-world dataset.

```python
import pandas as pd
import numpy as np

student_data = {
    "Name": ["Alice","Bob","Charlie","David","Eva","Frank","Grace","Helen"],
    "Age": [20,np.nan,22,21,np.nan,23,20,22],
    "Marks": [88,75,np.nan,91,67,np.nan,84,79],
    "Department": ["CS","IT",np.nan,"ECE","CS","IT","ECE",np.nan]
}

df = pd.DataFrame(student_data)
print(df)
```

# 📖 Part 1 – Understanding Missing Values

**Objective:** Identify missing values and understand how incomplete information appears inside a DataFrame.

## Tasks

Using the Practice Dataset:

- Identify the missing values.
- Count the missing values in each column.
- Find the column containing the highest number of missing values.

```python
#Identify which columns contain missing values.
#print("columns with missing values : \n",df.isna().any())
cols_with_missing = df.columns[df.isna().any()].tolist()
print("columns with missing values : ",cols_with_missing)
#Count missing values in each column.
count_of_cols_with_missing = df[cols_with_missing].isna().sum(axis=0)
print("Count of columns with missing values in each column : \n",count_of_cols_with_missing)
#Identify the column with the highest number of missing values.
print("column with the highest number of missing values : ",cols_with_missing[count_of_cols_with_missing.argmax()])
#Compare the total number of missing values with the total number of records.
print(f"the total number of missing values : {count_of_cols_with_missing.sum()}; the total number of records : {df.size}")
```

### 📝 Observation

- The **Age** column contains the highest number of missing values in the Practice Dataset.
- A total of **6 missing values** are present out of **32 data entries** (8 rows × 4 columns), indicating that approximately **18.75%** of the dataset contains missing information.
- Missing values are distributed across multiple columns rather than being concentrated in a single record.

# 📖 Part 3 – Removing Missing Values

**Objective:** Remove incomplete records using different variations of `dropna()`.

## ✅ Tasks

Experiment with:

- Remove rows containing missing values.
- Remove columns containing missing values.
- Remove rows only if every value is missing.
- Remove rows based on selected columns.

Compare the DataFrame shape after each operation.

```python
#Remove rows containing missing values.
print("Shape before operation : ",df.shape)
after_row_removal = df.dropna(axis=0)
print("Remove rows containing missing values : \n",after_row_removal)
print("Shape after operation : ",after_row_removal.shape)
```

```python
#Remove columns containing missing values.
print("Shape before operation : ",df.shape)
after_column_removal = df.dropna(axis=1)
print("Remove columns containing missing values : \n",after_column_removal)
print("Shape after operation : ",after_column_removal.shape)
```

```python
#Remove rows only if every value is missing.
print("Shape before operation : ",df.shape)
after_row_removal_only_if_missing = df.dropna(how="all")
print("Remove rows only if every value is missing : \n",after_row_removal_only_if_missing)
print("Shape after operation : ",after_row_removal_only_if_missing.shape)
```

```python
#Remove rows based on selected columns.
print("Shape before operation : ",df.shape)
after_row_removal_by_columns = df["Age"].dropna(how="all")
print("Remove rows only if every value is missing : \n",after_row_removal_by_columns)
print("Shape after operation : ",after_row_removal_by_columns.shape)
```

### 📝 Observation

- `dropna()` removes incomplete records based on the specified condition.
- Different parameters of `dropna()` can produce significantly different outputs, making parameter selection important.
- Comparing the DataFrame shape before and after `dropna()` clearly shows how much data is lost during the cleaning process.

# 📖 Part 4 – Filling Missing Values

**Objective:** Replace missing values using different imputation techniques.

## ✅ Tasks

Practice:

- Replace with a constant value.
- Replace with mean.
- Replace with median.
- Replace with mode.
- Forward Fill (`ffill`)
- Backward Fill (`bfill`)

Compare the resulting DataFrames after each method.

```python
#Replace with a constant value.
data_fill_with_constant = df.fillna(1)
print("Replace with a constant value : ",data_fill_with_constant)
```

```python
#Replace with mean.
#print("Replace with mean : \n",df.fillna(df["Age"].mean()))

# This fills every numeric column's missing values with that specific column's mean
print("Replace with mean : \n",df.fillna(df.mean(numeric_only=True)))
```

```python
#Forward Fill (`ffill`)
print("Forward Fill : \n",df.ffill())
#Backward Fill (`bfill`)
print("\nBackward fill : \n",df.bfill())
```

### 📝 Observation

- `fillna()` preserves the original structure of the dataset while replacing missing values.
- Different filling strategies (mean, median, mode, constant values, forward fill, and backward fill) produce different results depending on the data type and distribution.
- Choosing an appropriate filling method is essential to avoid introducing bias into the dataset.
