# 📂 Practice Dataset

Today's concepts will first be practiced using a small employee dataset where the results of grouping and aggregation are easy to understand and verify.

The Titanic dataset will be used only during the Mini Challenge and Bonus Challenge.

```python
import pandas as pd

employee_data = {
    "Employee":["Alice","Bob","Charlie","David","Eva","Frank","Grace","Helen","Ivan","Julia"],
    "Department":["HR","IT","IT","Finance","HR","Finance","IT","HR","Finance","IT"],
    "Gender":["F","M","M","M","F","M","F","F","M","F"],
    "Salary":[45000,70000,65000,90000,48000,85000,72000,50000,95000,68000],
    "Experience":[3,10,7,15,2,12,5,6,18,8]
}

practice_df = pd.DataFrame(employee_data)

print(practice_df)
```

# 📖 Part 1 – Understanding GroupBy

**Objective:** Learn how similar records can be grouped together before performing calculations.

## ✅ Tasks

Using the Practice Dataset:

- Group employees by Department.
- Display the grouped object.
- Iterate through each group.
- Display every department separately.

Observe how Pandas divides the dataset into independent groups.

```python
# Group employees by Department
grouped_departments = practice_df.groupby("Department")

# Display the grouped object (this prints the object type and memory location)
print("Group employees by Department : ", grouped_departments)

# Iterate through each group
for department_name, department_data in grouped_departments:
    
    # Display every department separately
    print(f"\n--- {department_name} Department ---")
    print(department_data)
```

### 📝 Observation

- `groupby()` divides the dataset into separate groups based on the values of one or more categorical columns.
- Each group behaves like an independent subset, allowing calculations to be performed on specific categories.
- Grouping data simplifies category-wise analysis and forms the foundation for aggregation in Pandas.

# 📖 Part 2 – Basic Aggregation

**Objective:** Apply aggregation functions to grouped data.

## ✅ Tasks

Find for each Department:

- Average Salary
- Maximum Salary
- Minimum Salary
- Total Salary
- Employee Count

```python
#Average Salary
for department_name, department_data in grouped_departments:
    
    # Display every department separately
    print(f"\n--- {department_name} Department ---")
    print("Average Salary for each Department : ",department_data["Salary"].mean())
#Maximum Salary
    print("Max Salary for each Department : ",department_data["Salary"].max())
#Minimum Salary
    print("Min Salary for each Department : ",department_data["Salary"].min())
#Total Salary
    print("Total Salary for each Department : ",department_data["Salary"].sum())
#Employee Count
    print("Employee Count : ",department_data["Salary"].count())
```

### 📝 Observation

- Individual aggregation functions such as `mean()`, `sum()`, `count()`, and `max()` summarize grouped data from different perspectives.
- Each aggregation provides unique insights, such as average performance, total contribution, or group size.
- Applying one aggregation at a time helps understand the purpose and output of each statistical function.

# 📖 Part 3 – Multiple Aggregations using `agg()`

**Objective:** Learn how to perform multiple aggregation functions on one or more columns using a single `agg()` operation.

## ✅ Tasks

Using the Practice Dataset:

Calculate the following for each Department using a single `agg()` call:

### Salary

- Mean
- Median
- Maximum
- Minimum

### Experience

- Mean
- Maximum

Observe how `agg()` allows multiple statistical summaries to be generated simultaneously.

```python
print("Salary and experience : \n",practice_df.groupby("Department").agg({
    "Salary":["mean","max","min","median"],
    "Experience":["mean","max"]
}))
```

### 📝 Observation

- `agg()` allows multiple statistical functions to be applied simultaneously within a single operation.
- Different columns can have different aggregation functions, making data summarization more efficient.
- Creating a consolidated summary table simplifies further analysis and reduces repetitive calculations.

# 📖 Part 4 – Grouping by Multiple Columns

**Objective:** Learn how to group data using multiple categorical columns to perform more detailed analysis.

## ✅ Tasks

Using the Practice Dataset:

Group the data by:

- Department
- Gender

Calculate:

- Average Salary
- Employee Count

Observe how grouping by multiple columns creates more detailed summaries for each category combination.

```python
print("Grouping Department and Gender : \n",practice_df.groupby(["Department","Gender"]).agg({"Salary" : ["mean"], "Gender":["count"]}))
```

### 📝 Observation

- Grouping by multiple columns creates more detailed summaries for every category combination.
- Multi-level grouping helps uncover relationships that cannot be observed through single-column grouping.
- This approach is useful for comparing metrics across multiple dimensions such as department and gender or passenger class and survival status.

# 📖 Part 5 – Business Interpretation

**Objective:** Interpret aggregated results and convert statistical summaries into meaningful business insights.

## ✅ Tasks

Using the aggregated results obtained earlier, answer the following questions:

- Which department has the highest average salary?
- Which department has the highest total payroll?
- Which department has the largest workforce?
- Which department has the most experienced employees?

For each answer, write one sentence explaining what the result suggests from a business perspective.

```python
dept_summary = practice_df.groupby("Department").agg({
    "Salary":["mean","sum"],
    "Experience":["max","count"]
})
#Which department has the highest average salary?
print("department with highest average salary : ",dept_summary["Salary"]["mean"].idxmax())
#Which department has the highest total payroll?
print("department with highest total payroll : ",dept_summary["Salary"]["sum"].idxmax())
#Which department has the largest workforce?
print("department with largest workforce : ",dept_summary["Experience"]["count"].idxmax())
#Which department has the most experienced employees?
print("department with most experienced employees : ",dept_summary["Experience"]["max"].idxmax())
```

### 📝 Observation

- Aggregated results become valuable only when they are interpreted in a business context.
- Comparing summary statistics helps identify trends, strengths, and potential areas for improvement.
- Business insights should explain what the numbers imply rather than simply repeating the calculated values.

# 🧩 Mini Challenge – Titanic Business Report

## Scenario

Your manager wants a quick summary of the Titanic passengers before starting detailed analysis.

Using only the Titanic dataset, prepare a report containing:


---

### Passenger Class Analysis

- Average passenger age by passenger class.
- Average fare by passenger class.
- Passenger count by passenger class.

```python
df = pd.read_csv(r"C:\Users\manid\OneDrive\Desktop\Project-Atlas\Datasets\Titanic\train.csv")

pclass_summary = df.groupby(["Pclass"]).agg({"Age" : ["mean"], "Fare" : ["mean"], "PassengerId" : ["count"]})
#Average passenger age by passenger class.
print("Average passenger age by passenger class : \n",pclass_summary["Age"]["mean"])
#Average fare by passenger class.
print("\nAverage fare by passenger class : \n",pclass_summary["Fare"]["mean"])
#Passenger count by passenger class.
print("\nPassenger count by passenger class : \n",pclass_summary["PassengerId"]["count"])
```

### Survival Analysis

- Survival count by gender.
- Survival count by passenger class.

---

```python
survival_summary_by_gender = df.groupby(["Sex"]).agg({"Survived" : ["value_counts"]})
#Survival count by gender.
print("Survival count by gender : \n",survival_summary_by_gender)
#Survival count by passenger class.
survival_summary_by_pclass = df.groupby(["Pclass"]).agg({"Survived" : ["value_counts"]})
print("\nSurvival count by passenger class : \n",survival_summary_by_pclass)
```

### Embarkation Analysis

- Passenger count by embarkation port.
- Average fare by embarkation port.

```python
#Passenger count by embarkation port.
passenger_summary_by_embarkation_port = df.groupby(["Embarked"]).agg({"PassengerId" : ["count"], "Fare" : ["mean"]})
print("Passenger count by embarkation port : \n",passenger_summary_by_embarkation_port["PassengerId"]["count"])
#Average fare by embarkation port.
print("\nAverage Fare by embarkation port : \n",passenger_summary_by_embarkation_port["Fare"]["mean"])
```

## 📌 Mini Challenge Summary

Using the Titanic dataset, I grouped passengers by categories such as passenger class, gender, and embarkation port to generate meaningful summaries.

The analysis highlighted differences in passenger distribution, average fares, survival counts, and age across various groups. These aggregated results provided a clear overview of the dataset and demonstrated how grouping can transform raw data into concise, decision-oriented information for further exploratory analysis.
