# 📖 Part 6 – Filtering Data

**Objective:** Filter records using logical conditions to answer analytical questions.

## ✅ Tasks

Apply filtering to:

- Select passengers based on age.
- Select passengers based on gender.
- Combine multiple conditions.
- Use `isin()`.
- Use `between()`.
- Use `query()`.

```python
#Select passengers based on age.
mask = df.where(df["Age"]>70)
print("Passengers age > 70 : \n",mask.dropna())
#Select passengers based on gender.
mask_gender = df.where(df["Sex"]=="male")
print("Passengers that are male : \n",mask_gender.dropna())
```

```python
#Combine multiple conditions.
mask_1 = df.where((df["Age"]>50)&(df["Sex"]=="female")).dropna()
print("Passengers whose age > 50 and gender is female : \n",mask_1)
```

```python
#Use `isin()`.
print("Passengers in A5 cabin : \n",df[df["Cabin"].isin(["A5"])])
#Use `between()`.
print("Passengers in between age 60 to 70 : \n",df[df["Age"].between(60,70)])
```

```python
#Use `query()`.
print("Passengers age > 70 : \n",df.query("Age>75"))
```

### 📝 Observation

- Filtering reduces the dataset to only relevant records.
- Multiple conditions allow more precise analysis.
- Different filtering methods improve flexibility and readability.

# 🧩 Analyst Challenge – Passenger Profile Investigation

## Scenario

You have been asked to prepare a quick passenger profile report before the analytics team begins working on the Titanic dataset.

Without modifying the original dataset, answer the following questions.

---

## Passenger Analysis

Determine:

- How many passengers were travelling in each passenger class?
- How many passengers were male and female?
- How many passengers survived and how many did not?
- Which passenger paid the highest fare?
- Which passenger paid the lowest fare?
- What is the average age of passengers?
- Which passenger was the oldest?
- Which passenger was the youngest?

---

```python
#Passengers travelling in each passenger car

# value_counts() returns a Series with the unique values as the index and counts as the values
# sort_index() ensures they are ordered (e.g., Class 1, 2, 3) just like np.unique does
passenger_counts = df["Pclass"].value_counts().sort_index()
print(passenger_counts)
```

```python
#No of female and male passengers
# value_counts() returns a Series with the unique values as the index and counts as the values
# sort_index() ensures they are ordered (e.g., Class 1, 2, 3) just like np.unique does
gender_counts = df["Sex"].value_counts()
print("gender wise count : ",gender_counts)
```

```python
#No of passengers survived and not survived

df["Survived"] = df["Survived"].replace({0:"DEAD",1:"SURVIVED"})
death_count = df["Survived"].value_counts()
print(death_count)
```

```python
#Passenger who paid highest fare
print("Passenger who paid highest fare : \n",df.iloc[df["Fare"].argmax()])
```

```python
#Passenger who paid Lowest fare
print("Passenger who paid Lowest fare : \n",df.iloc[df["Fare"].argmin()])

#Average age of passengers
print("\nAverage age of passengers : ",df["Age"].mean())

#Oldest passenger
print("\nThe oldest passenger is : \n",df.iloc[df["Age"].argmax()])
```

```python
#Youngest passenger
print("\nThe youngest passenger is : \n",df.iloc[df["Age"].argmin()])
```

## Data Exploration

Find:

- All passengers older than 60 years.
- All female passengers travelling in First Class.
- All passengers younger than 15 years who survived.
- All passengers who paid more than the average fare.
- All passengers who embarked from Southampton (`S`) and travelled in First Class.

---

```python
#All passengers older than 60
print("All Passengers older than 60 : \n",df.where(df["Age"]>60).dropna())

#All female passengers travelling in 1st class
print("\nFemale passengers travelling in 1st class : \n",df.query("Sex == 'female' and Pclass == 1"))
```

```python
#All passengers younger than 15 and survived
print("passengers younger than 15 and survived : \n",df.query("Age<15 and Survived == 'SURVIVED'"))
```

```python
#All passengers who paid more than average fare
avg_fare = df.Fare.mean()
print("passengers who paid more than average fare : \n",df.query("Fare > @avg_fare"))
```

```python
#All passengers who embarked from Southampton (`S`) and travelled in First Class
print("Passengers who embarked from Southampton (`S`) and travelled in First Class : \n",df.query("Embarked == 'S' and Pclass == 1"))
```
