# 📂 Practice Dataset

Today's concepts will first be practiced using a small dataset where sorting, counting, and identifying unique values can be easily verified.

The same analytical thinking will later be applied to the Titanic dataset during the Mini and Bonus Challenges.

```python
import pandas as pd

employee_data = {
    "Employee": ["Alice","Bob","Charlie","David","Eva","Frank","Grace","Helen"],
    "Department": ["HR","IT","IT","Finance","HR","Finance","IT","HR"],
    "Age": [28,35,31,42,25,38,29,33],
    "Salary": [45000,70000,65000,90000,48000,85000,72000,50000],
    "Experience": [3,10,7,15,2,12,5,6]
}

df1 = pd.DataFrame(employee_data)

print(df1)
```

# 📖 Part 1 – Sorting Data

**Objective:** Learn how to organize data in ascending and descending order using Pandas.

## ✅ Tasks

Practice:

- Sort by Age (Ascending)
- Sort by Age (Descending)
- Sort by Salary
- Sort by Department
- Sort by Department and Salary together

```python
#Sort by Age (Ascending)
print("Sort by Age (Ascending) : \n",df1.sort_values(by="Age"))
#Sort by Age (Descending)
print("\nSort by Age (Descending) : \n",df1.sort_values(by="Age",ascending=False))
#Sort by Salary
print("\nSort by Salary : \n",df1.sort_values(by=["Salary","Department"],ascending=False))
```

```python
#Sort by Department
print("\nSort by Department : \n",df1.sort_values(by=["Department"]))

#Sort by Department and Salary together
print("\nSort by Department and Salary : \n",df1.sort_values(by=["Salary","Department"],ascending=False))
```

### 📝 Observation

- `sort_values()` organizes the dataset based on one or more column values without permanently modifying the original DataFrame unless explicitly assigned.
- Sorting can be performed in both ascending and descending order, making it easier to identify the highest and lowest values.
- Multiple-column sorting allows hierarchical organization of data, which is useful when analyzing records with similar categories.

# 📖 Part 2 – Sorting by Index

**Objective:** Understand how DataFrame indexes can also be sorted independently of column values.

## ✅ Tasks

Practice:

- Sort the index in ascending order
- Sort the index in descending order

Observe how index sorting differs from value sorting.

```python
#Sort by index in ascending order
print("\nSort by index in ascending order : \n",df1.sort_index())

#Sort by index in descending order
print("\nSort by index in descending order : \n",df1.sort_index(ascending=False))
```

### 📝 Observation

- `sort_index()` rearranges the DataFrame based on index labels rather than column values.
- Index sorting is useful when the row order has changed after filtering, shuffling, or concatenation.
- Unlike `sort_values()`, `sort_index()` does not consider the values stored inside the columns.

# 📖 Part 3 – Unique Values

**Objective:** Identify distinct values present within categorical columns.

## ✅ Tasks

Practice:

- Find unique departments.
- Count the number of unique departments.
- Find unique experience values.
- Compare `unique()` and `nunique()`.

```python
#Find unique departments.
print("Find the unique departments : ",df1["Department"].unique())
#Count the number of unique departments.
print("Count the number of unique departments : ",df1["Department"].unique().size)
#Find unique experience values.
print("Find unique experience values : ",df1["Experience"].unique())
#Compare `unique()` and `nunique()`.
print("Compare `unique()` and `nunique()` : ",df1["Experience"].nunique()) #gives the number of unique elements
```

### 📝 Observation

- `unique()` returns all distinct values present in a column.
- `nunique()` returns only the count of unique values instead of the values themselves.
- Identifying unique values helps understand categorical features before further analysis.

# 📖 Part 4 – Frequency Analysis

**Objective:** Measure how frequently different values occur within a dataset.

## ✅ Tasks

Practice:

- Count employees in each department.
- Count employees by experience.
- Count salaries (if duplicates exist).
- Display frequencies in descending order.

```python
#Count employees in each department.
print("Count employees in each department : ",df1["Department"].value_counts())
#Count employees by experience.
print("Count employees by experience : ",df1["Experience"].value_counts())
#Count salaries (if duplicates exist).
print("Count salaries (if duplicates exist) : ",df1["Salary"].nunique())
#Display frequencies in descending order.
print("Display frequencies in descending order : ",df1["Department"].value_counts().sort_values(ascending=False))
```

### 📝 Observation

- `value_counts()` quickly summarizes how frequently each value appears within a column.
- Frequency analysis helps identify dominant categories and understand data distribution.
- By default, the results are displayed in descending order of frequency.

# 📖 Part 5 – Ranking & Quick Exploration

**Objective:** Combine sorting and counting techniques to quickly explore a dataset.

## ✅ Tasks

Determine:

- Highest paid employee
- Lowest paid employee
- Most experienced employee
- Youngest employee
- Oldest employee

```python
#Highest paid employee
print("Highest paid employee : \n",df1.iloc[df1["Salary"].argmax()])
#Lowest paid employee
print("\nLowest paid employee : \n",df1.iloc[df1["Salary"].argmin()])
#Most experienced employee
print("\nMost experienced employee : \n",df1.iloc[df1["Experience"].argmax()])
```

```python
#Youngest employee
print("Youngest employee : \n",df1.iloc[df1["Age"].argmin()])
#Oldest employee
print("\nOldest Employee :\n",df1.iloc[df1["Age"].argmax()])
```

### 📝 Observation

- Combining sorting and filtering makes it easy to identify important records such as highest salaries or oldest employees.
- Ranking allows analysts to quickly answer business questions without performing complex calculations.
- Data exploration becomes more efficient when multiple Pandas operations are combined logically.

# 🚀 Bonus Challenge – Passenger Ranking Report

Prepare a ranked report using the Titanic dataset.
