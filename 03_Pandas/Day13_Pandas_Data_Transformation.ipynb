# 📂 Practice Dataset

```python
import pandas as pd
employee_data = {
    "Employee":["Alice","Bob","Charlie","David","Eva","Frank","Grace","Helen"],
    "Department":["HR","IT","IT","Finance","HR","Finance","IT","HR"],
    "Gender":["F","M","M","M","F","M","F","F"],
    "Salary":[45000,70000,65000,90000,48000,85000,72000,50000],
    "Experience":[3,10,7,15,2,12,5,6]
}

practice_df = pd.DataFrame(employee_data)

print(practice_df)
```

# 📖 Part 1 – Transforming Values with `map()`

**Objective:** Learn how to map existing values to new values using a dictionary.

## ✅ Tasks

Using the Practice Dataset:

- Convert Gender from `M` and `F` to `Male` and `Female`.
- Assign numeric IDs to each Department.
- Create a new column using mapped department codes.

Observe how `map()` transforms values without changing the underlying structure of the dataset.

```python
#Convert Gender from `M` and `F` to `Male` and `Female`.
practice_df["Gender"] = practice_df["Gender"].map({"M" : "Male", "F" : "Female"})
print(practice_df)
#Assign numeric IDs to each Department.
practice_df["DepartmentCodes"] = practice_df["Department"].map({"HR" : "1", "IT" : "2", "Finance" : "3"})
#Create a new column using mapped department codes.
print("\n New Department Codes : \n",practice_df)
```

### 📝 Observation

- `map()` transforms existing values into new values using a mapping dictionary or function.
- It is best suited for converting values within a single Series while preserving the original data structure.
- Mapping improves data consistency and makes categorical information easier to interpret or analyze.

# 📖 Part 2 – Replacing Values with `replace()`

**Objective:** Learn how to replace one or more existing values throughout a DataFrame or specific columns.

## ✅ Tasks

Using the Practice Dataset:

- Replace `HR` with `Human Resources`.
- Replace `IT` with `Information Technology`.
- Replace multiple department names using a dictionary.
- Replace one salary value with an updated amount.

Observe how `replace()` modifies existing values directly.

```python
#Replace `HR` with `Human Resources`.
practice_df = practice_df.replace("HR","Human Resources")
print("Replace `HR` with `Human Resources` : \n",practice_df)
#Replace `IT` with `Information Technology`.
practice_df = practice_df.replace("IT","Information Technology")
print("\nReplace `IT` with `Information Technology` : \n",practice_df)
#Replace multiple department names using a dictionary.
practice_df["Department"] = practice_df["Department"].replace({"Information Technology" : "IT", "Finance" : "Fin"})
print("\nReplace multiple department names using a dictionary : \n",practice_df)
```

```python
#Replace one salary value with an updated amount.
practice_df = practice_df.replace(45000 , 60000)
print("\nReplace one salary value with an updated amount : \n",practice_df)
```

### 📝 Observation

- `replace()` substitutes one or more existing values with new values across a Series or DataFrame.
- Multiple replacements can be performed efficiently using a dictionary.
- Standardizing inconsistent values improves data quality and readability before analysis.

# 📖 Part 3 – Custom Transformations with `apply()`

**Objective:** Apply custom functions to transform data using `apply()`.

## ✅ Tasks

Using the Practice Dataset:

- Create a Salary Grade column (A, B, C) based on salary.
- Categorize employees as Junior or Senior based on experience.
- Create a new Bonus column where bonus equals 10% of salary.

Observe how `apply()` enables custom business logic to be applied across rows or columns.

```python
#Create a Salary Grade column (A, B, C) based on salary.
# 1. Define the rules
def get_salary_grade(salary):
    if salary >= 80000:
        return "A"
    elif salary >= 60000:
        return "B"
    else:
        return "C"

# 2. Apply the function to the Salary column
practice_df["Salary Grade"] = practice_df["Salary"].apply(get_salary_grade)

print(practice_df)

#Categorize employees as Junior or Senior based on experience.
practice_df["Employee Category"] = practice_df["Experience"].apply(lambda x : "Senior" if x>9 else "Junior")
print("\nCategorize employees as Junior or Senior based on experience : \n",practice_df)
```

```python
#Create a new Bonus column where bonus equals 10% of salary.
practice_df["Bonus"] = practice_df["Salary"].apply(lambda x : x*(0.1))
print("\n Create a new Bonus column where bonus equals 10% of salary \n",practice_df)
```

### 📝 Observation

- `apply()` executes custom functions on Series or DataFrames to perform flexible transformations.
- Business rules can be implemented easily by combining conditional logic with `apply()`.
- It is especially useful when built-in Pandas functions cannot directly solve the problem.

# 📖 Part 4 – Row-wise Transformations with `apply()`

**Objective:** Learn how to apply custom functions that use multiple columns from the same row.

## ✅ Tasks

Using the Practice Dataset:

Create the following columns using `apply(axis=1)`:

- Employee Level
    - Junior
    - Mid-Level
    - Senior

Use both Salary and Experience.

---

```python
def employee_level(row):
    salary = row["Salary"]
    experience = row["Experience"]
    if salary>50000 and experience>5 :
        return "Senior"
    elif salary<50000 or experience<5:
        return "Junior"
    else:
        return "Midlevel"

practice_df["Employee Level"] = practice_df.apply(employee_level,axis=1)
print(practice_df)
```

---

Create a Performance Bonus column.

Business Rule:

- Experience > 10 years AND Salary > 80,000 → 20%
- Experience > 5 years → 10%
- Otherwise → 5%

---

```python
def performance_bonus_claim(row):
    experience = row["Experience"]
    salary = row["Salary"]
    if experience > 10 and salary > 80000:
        return salary*(0.2)
    elif experience > 5:
        return salary*(0.1)
    else:
        return salary*(0.05)
practice_df["Bonus"] = practice_df.apply(performance_bonus_claim,axis=1)
print(practice_df)
```

---

Create an Employee Category:

- High Performer
- Experienced
- Developing

based on Salary and Experience together.

Observe how `apply(axis=1)` allows business rules to use multiple columns simultaneously.

---

```python
def employee_category(row):
    salary = row["Salary"]
    experience = row["Experience"]
    if salary>75000 and experience>8 :
        return "High Performer"
    elif experience>11:
        return "Experienced"
    else:
        return "Developing"

practice_df["Employee Category"] = practice_df.apply(employee_category,axis=1)
print(practice_df)
```

### 📝 Observation

- `apply(axis=1)` processes one complete row at a time, allowing multiple columns to be used within a single transformation.
- Row-wise transformations make it possible to create new features based on business logic involving multiple attributes.
- This approach is commonly used during feature engineering and data preparation for analytics and machine learning.

```python
#Encode `Sex` into numeric values.
df["Sex"] = df["Sex"].map({"Male" : 0, "Female" : 1})
print("Encode `Sex` into numeric values (0,1) : \n",df)
```

```python
#Encode Passenger Class into descriptive labels.
df["Pclass"] = df["Pclass"].map({1 : "First Class" , 2 : "Second Class", 3:"Third Class"})
print("Encode Passenger Class into descriptive labels : \n",df)
```

```python
#Create a Family Size category using `SibSp` and `Parch`.
def family_size(row):
    sibsp = row["SibSp"]
    parch = row["Parch"]
    if (sibsp+parch)<=2:
        return "Nuclear"
    elif (sibsp+parch) in range(3,6):
        return "Small"
    elif (sibsp+parch)>5:
        return "Joint"
df["Family Size"] = df.apply(family_size,axis=1)
print("Create a Family Size category using `SibSp` and `Parch` : \n",df)
```

```python
#Create a Fare Level column.
def fare_category(fare):
    if fare < 10:
        return "Budget"
    elif fare in range(11,25):
        return "Standard"
    elif fare>25:
        return "Premium"
df["Fare Category"] = df["Fare"].apply(fare_category)
print(df)
```
