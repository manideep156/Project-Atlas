# 📖 Part 1 – Introduction to Pandas

## What is Pandas?

Pandas is an open-source Python library designed for data manipulation and analysis.

It provides easy-to-use data structures for working with structured or tabular data.

Pandas is built on top of **NumPy**, combining NumPy's computational efficiency with labels, indexing, and many tools for handling real-world datasets.

---

## Why Do We Need Pandas?

Although NumPy is excellent for numerical computations, it has several limitations when working with real datasets.

Pandas solves these problems by providing:

- Named rows and columns
- Easy filtering
- Data cleaning
- Missing value handling
- Reading data from files
- Statistical summaries
- Data grouping and aggregation

---

## Real-World Applications

Pandas is widely used for:

- Data Analysis
- Business Intelligence
- Financial Analysis
- Machine Learning
- Data Cleaning
- Exploratory Data Analysis (EDA)

# 📖 Part 3 – Pandas Series

## What is a Series?

A Series is a **one-dimensional labeled array** capable of storing data of almost any type.

Each value in a Series has an associated **index**, making it easier to access and manipulate data.

A Series can be created from:

- Python Lists
- NumPy Arrays
- Dictionaries
- Scalar Values

## Practice

Create Series using:

Observe:

```python
import numpy as np
import pandas as pd
Marks=[82, 91, 76, 88, 95]

list_series = pd.Series(Marks)
print("Series using List : \n",list_series)

numpy_array = np.arange(1,4)
numpy_array_series = pd.Series(numpy_array)
print("Series using Numpy Array : \n",numpy_array_series)

Dictionary = {"Apple" : 220, "Banana" : 45, "Orange" : 110, "Mango" : 180}
dictionary_series = pd.Series(Dictionary)
print("Series using Dictionary : \n",dictionary_series)

Cities = ["Hyderabad", "Visakhapatnam", "Bengaluru", "Chennai", "Pune"]
custom_index = pd.Series(Cities, index=["TS","AP","KS","TN","MH"])
print("Series with custom index : \n",custom_index)
```

A DataFrame is a **two-dimensional labeled data structure** consisting of rows and columns.

It is the most commonly used Pandas object because it resembles spreadsheets, SQL tables, and Excel sheets.

Each column can contain different data types.

```python
data = {"Day":[1,2,3,4,5,6,7,8],
        "Topic": ["Python Revision","NumPy Basics","Broadcasting","Advanced NumPy","NumPy Mastery","Portfolio Project","Revision","Pandas Fundamentals"],
        "Hours_Studied": [3.5,5.0,5.2,4.8,5.5,6.2,3.0,4.0],
        "Problems_Solved" : [12,18,22,20,25,30,10,15],
        "GitHub_Commit":["Yes","Yes","Yes","Yes","Yes","Yes","Yes","No"],
        "Status":["Completed","Completed","Completed","Completed","Completed","Completed","Completed","In Progress"]}

df=pd.DataFrame(data)
print("DataFrame : \n",df)

#Display the first five rows.
print("\nfirst five rows : \n",df.head(5))
```

## Column Selection

Display:

- Topic
- Hours_Studied
- Problems_Solved

```python
#Topic
print("Topics : \n",df["Topic"])
#Hours_Studied
print("\nHours_Studied\n",df["Hours_Studied"])
#Problems_Solved
print("Problems Solved : ",df["Problems_Solved"])
```

## Basic Analysis

Find:

- Total study hours.
- Average study hours.
- Total problems solved.
- Average problems solved per day.

```python
#Total study hours.
print("Total study hours : ",df["Hours_Studied"].sum())
#Average study hours.
print("Average study hours : ",df["Hours_Studied"].mean())
#Total problems solved.
print("Total problems solved : ",df["Problems_Solved"].sum())
#Average problems solved per day.
print("Average problems solved per day : ",df["Problems_Solved"].mean())
```

## Productivity Analysis

Identify:

- Day with the maximum study hours.
- Day with the minimum study hours.
- Day with the maximum problems solved.

```python
#Day with the maximum study hours.
print("Day with the maximum study hours : ",df["Hours_Studied"].argmax() +1)
#Day with the minimum study hours.
print("Day with the minimum study hours : ",df["Hours_Studied"].argmin() +1)
#Day with the maximum problems solved.
print("Day with the maximum problems solved : ",df["Problems_Solved"].argmax() +1)
```

## Filtering

Display:

- Days where study hours exceeded **5 hours**.
- Days where more than **20 problems** were solved.
- Days with **Completed** status only.

```python
#Days where study hours exceeded **5 hours**.
mask=df.where(df["Hours_Studied"]>5)
print("Days where study hours exceeded 5 hours : \n",mask.dropna())
#Days where more than **20 problems** were solved.
mask_2 = df.where(df["Problems_Solved"]>20)
print("\nDays where more than 20 problems were solved : \n",mask_2.dropna())
#Days with **Completed** status only.
mask_3 = df.where(df["Status"]=="Completed")
print("Days with Completed status only : \n",mask_3.dropna())
```
