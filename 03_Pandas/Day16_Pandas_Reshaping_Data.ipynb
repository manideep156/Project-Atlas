# 📂 Practice Dataset

Today's concepts will be practiced using a monthly sales dataset stored in a wide format.

The goal is to transform this dataset into different layouts that are more suitable for analysis and visualization.

```python
import pandas as pd

sales_data = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet",
        "Headphones"
    ],

    "January": [
        120,
        180,
        95,
        140
    ],

    "February": [
        135,
        170,
        105,
        150
    ],

    "March": [
        145,
        190,
        110,
        155
    ]
}

practice_df = pd.DataFrame(sales_data)
```

# 📖 Part 2 – Reshaping with `melt()`

**Objective:** Learn how to convert a wide-format dataset into a long-format dataset.

## ✅ Tasks

Using the Practice Dataset:

- Convert the monthly sales columns into rows using `melt()`.
- Keep the Product column as the identifier.
- Rename the generated columns to meaningful names.
- Compare the original and reshaped datasets.
- Explain why the long format is useful for further analysis.

Focus on understanding how `melt()` restructures data without changing its values.

```python
melt1 = pd.melt(practice_df,id_vars="Product",value_name="Sales",var_name="Month")
print(practice_df)
print("\n",melt1)
```

### 📝 Observation

- `melt()` converts wide-format data into long-format data without changing the underlying values.
- Identifier columns remain unchanged while measurement columns are transformed into rows.
- Long-format data is easier to analyze, visualize, and use in machine learning workflows.

# 📖 Part 3 – Reshaping with `stack()` & `unstack()`

**Objective:** Learn how to reshape data by converting columns into index levels and vice versa.

## ✅ Tasks

Using the Practice Dataset:

- Set Product as the index.
- Apply `stack()` and observe the output.
- Convert the stacked data back using `unstack()`.
- Compare the original DataFrame with the restored version.
- Explain how stack and unstack differ from `melt()`.

Focus on understanding index-based reshaping.

```python
practice_df = practice_df.set_index("Product")
stacked = practice_df.stack()
print(stacked)
unstacked = stacked.unstack()
print("\n",unstacked)
```

### 📝 Observation

- `stack()` converts columns into index levels, while `unstack()` performs the reverse operation.
- These methods reshape data using the DataFrame index rather than creating new columns.
- Stack and unstack are especially useful when working with MultiIndex DataFrames and hierarchical data.

# 📖 Part 4 – Expanding List-Like Data with `explode()`

**Objective:** Learn how to separate list-like values into individual rows using `explode()`.

```python
orders = pd.DataFrame({
    "Order_ID":[101,102,103],
    "Customer":["Alice","Bob","Charlie"],
    "Products":[
        ["Laptop","Mouse"],
        ["Phone"],
        ["Tablet","Cover","Pen"]
    ]
})

print(orders)
print("\n",orders.explode("Products"))
```

### 📝 Observation

- `explode()` separates list-like values into individual rows while preserving the remaining columns.
- Expanding nested data simplifies filtering, grouping, and aggregation.
- Many real-world datasets require exploding before meaningful analysis can be performed.

# 🧩 Mini Challenge – Student Marks Reshaping

## Scenario

A school's marks dataset stores subject scores in separate columns.

Create the following dataset:

- Student Name
- Mathematics
- Science
- English

Using Pandas:

- Convert the dataset from wide format to long format using `melt()`.
- Rename the generated columns appropriately.
- Identify the highest score for each subject.
- Calculate the average score for each subject.
- Sort the reshaped dataset by marks in descending order.

Write a short summary explaining why the reshaped dataset is more suitable for analysis.

```python
data = {
    "Student Name": [
        "Mani",
        "Poorna",
        "Rithvik",
        "Anirudh"
    ],

    "Maths": [
        70,
        80,
        95,
        60
    ],

    "Science": [
        35,
        70,
        55,
        50
    ],

    "English": [
        45,
        80,
        90,
        95
    ]
}


stu_df = pd.DataFrame(data)
print(stu_df)
melt2 = pd.melt(stu_df,id_vars="Student Name",value_name="Marks",var_name="Subject")
print("\n",melt2)
print("\nHighest score of each subject : \n",melt2.groupby(by="Subject")["Marks"].max())
print("\nAverage score of each subject : \n",melt2.groupby(by="Subject")["Marks"].mean())
```

```python
print("\nSorting melted df by marks : \n",melt2.sort_values("Marks",ascending=False))
```

## 📌 Mini Challenge Summary

Using a student marks dataset, I converted subject-wise marks stored in separate columns into a long-format dataset using `melt()`. I then analyzed subject-wise performance by identifying the highest scores, calculating average marks, and sorting the reshaped data.

# 🚀 Bonus Challenge – Monthly Sales Transformation Dashboard

## Scenario

A retail company stores monthly sales in separate columns.

Create a dataset containing:

- Product
- January
- February
- March
- April

Using Pandas:

- Convert the dataset into a long format.
- Identify the month with the highest sales for each product.
- Calculate the average monthly sales for each product.
- Create a pivot table to reconstruct the original wide-format report.
- Verify that the reshaped data can be converted back without losing information.

Finally,

Write **five analytical insights** explaining how reshaping simplifies reporting, visualization, and future analysis.

```python
data1 = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet",
        "Headphones"
    ],

    "January": [
        120,
        180,
        95,
        140
    ],

    "February": [
        135,
        170,
        105,
        150
    ],

    "March": [
        145,
        190,
        110,
        155
    ],
    "April": [
        225,
        120,
        80,
        115
    ]
}

df = pd.DataFrame(data1)


melt3 = pd.melt(df,id_vars="Product",var_name="Month",value_name="Sales")
print(melt3)
print("\nMonth with highest sales : ",melt3.groupby(by="Month")["Sales"].sum().idxmax())
print("\nAverage monthly sales : \n",melt3.groupby(by="Month")["Sales"].mean())
pivot = pd.pivot(melt3,index="Product",columns=["Month"],values="Sales")
print("\n",pivot)
```

```python

print("\n", pd.melt(pivot.reset_index(), id_vars="Product", var_name="Month", value_name="Sales"))
```

## 💡 Analytical Insights

1. Long-format datasets simplify month-wise trend analysis because each sales observation is represented as an individual row.

2. Reshaping data with `melt()` prepares datasets for visualization libraries such as Seaborn and Plotly, which generally expect tidy data.

3. Pivot tables can reconstruct summarized reports from reshaped data without losing information, demonstrating the reversibility of many reshaping operations.

4. Separating product and monthly sales into individual observations enables easier filtering, grouping, and aggregation across different business dimensions.

5. Choosing the appropriate reshaping technique improves data organization and creates datasets that are better suited for reporting, dashboard development, and machine learning workflows.
