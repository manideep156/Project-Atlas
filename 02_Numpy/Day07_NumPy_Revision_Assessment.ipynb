## Task 1 – Dataset Exploration

Find:

- Shape

## Task 2 – Product-Wise Analysis

Calculate:

- Product-wise totals
- Product-wise averages
- Product-wise maximum values
- Product-wise minimum values
- Product-wise standard deviations


````python
#Product-wise totals
Product_wise_totals = np.sum(sales,axis=0)
print("Product-wise totals : ",Product_wise_totals)
#Product-wise averages
Product_wise_averages = np.mean(sales,axis=0)
print("Product-wise Averages : ",Product_wise_averages)
#Product-wise maximum values
Product_wise_max = np.max(sales,axis=0)
print("Product-wise max : ",Product_wise_max)
#Product-wise minimum values
Product_wise_min = np.min(sales,axis=0)
print("Product-wise minimum : ",Product_wise_min)
#Product-wise standard deviations
Product_wise_std = np.std(sales,axis=0)
print("Product-wise std : ",Product_wise_std)
````

## Task 3 – Week-Wise Analysis

Calculate:

- Weekly totals
- Weekly averages
- Best-performing week
- Lowest-performing week

````python
#Weekly totals
Weekly_totals = np.sum(sales,axis=1)
print("Weekly totals : ",Weekly_totals)
#Weekly averages
Weekly_averages = np.mean(sales,axis=1)
print("Weekly Averages : ",Weekly_averages)
#Best-performing week
print("Best-performing week : ",np.argmax(Weekly_totals))
#Lowest-performing week
print("Lowest-performing week : ",np.argmin(Weekly_totals))
````

## Task 5 – Boolean Masking

Find:

- All sales values greater than 50
- All values below 40
- All values between 40 and 55 inclusive

````python
#All sales values greater than 50
print("sales values greater than 50 : ",sales[sales>50])
#All values below 40
print("All values below 40 : ",sales[sales<40])
#All values between 40 and 55 inclusive
print("All values between 40 and 55 inclusive : ",sales[(sales>=40)&(sales<=55)])
````

## Task 6 – Logical Row Analysis

Identify weeks where:

- Every product sold more than 35
- At least one product sold more than 55
- Every product sold at least 40

````python
#Every product sold more than 35
print("Every product sold more than 35 : ",np.where(np.all((sales>35),axis=1))[0])
#At least one product sold more than 55
print("At least one product sold more than 55 : ",np.where(np.any((sales>55),axis=1))[0])
#Every product sold at least 40
print("Every product sold at least 40 : ",np.where(np.all((sales>=40),axis=1))[0])
````

Create a NumPy summary table containing:

# ⚡ NumPy Quick Revision Guide

## Array Creation

- `np.array()` creates a NumPy array.
- `np.arange()` creates values using a start, stop, and step.
- `np.linspace()` creates evenly spaced values.
- `np.zeros()` and `np.ones()` create initialized arrays.

---

## Important Array Attributes

- `.shape` → dimensions of the array
- `.size` → total number of values
- `.ndim` → number of dimensions
- `.dtype` → data type
- `.nbytes` → total memory used

---

## Indexing and Slicing

- `arr[row, column]` accesses a specific 2D element.
- `arr[:, column]` selects a complete column.
- `arr[row, :]` selects a complete row.
- Slicing usually returns a view.
- Fancy indexing usually returns a copy.

---

## Reshaping

- `reshape()` changes dimensions without changing values.
- `flatten()` creates a 1D copy.
- `.T` transposes rows and columns.

---

## Aggregation and Axis

- `axis=0` performs operations column-wise.
- `axis=1` performs operations row-wise.

Common functions:

- `np.sum()`
- `np.mean()`
- `np.median()`
- `np.min()`
- `np.max()`
- `np.std()`
- `np.var()`

---

## Boolean Masking

Boolean masking filters data using conditions.

Important rules:

- Use `&` instead of `and`
- Use `|` instead of `or`
- Use `~` for NOT
- Wrap each condition in parentheses

---

## Broadcasting

Broadcasting allows operations between arrays with compatible shapes.

It is useful for:

- Applying product-specific increases
- Simulating promotions
- Adjusting columns
- Performing transformations without loops

---

## Searching and Ranking

- `np.where()` returns matching indices.
- `np.argmax()` returns the index of the maximum value.
- `np.argmin()` returns the index of the minimum value.
- `np.argsort()` returns sorting indices.
- `np.unique()` finds unique values and frequencies.

---

## Stacking

- `np.vstack()` adds rows.
- `np.hstack()` joins arrays horizontally.
- `np.column_stack()` converts 1D arrays into columns.

---

## Copy vs View

- A view shares memory with the original.
- A copy owns separate memory.
- Changes to a view may affect the original.
- Use `.copy()` when preserving the original data is important.
