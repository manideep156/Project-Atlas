# 📖 1. Fancy Indexing


## Dataset

Matrix A

```
[[11,12,13,14],
 [21,22,23,24],
 [31,32,33,34],
 [41,42,43,44]]
```

---

## 🎯 Tasks

Without using loops:

- Extract rows 1 and 3.
- Extract columns 2 and 4.
- Extract the elements:
  - 11
  - 23
  - 34
  - 42
- Extract the four corner elements.
- Reverse the row order using indexing.

---

````python
import numpy as np

arr = np.array([[11,12,13,14],
 [21,22,23,24],
 [31,32,33,34],
 [41,42,43,44]])
print(arr)
````

````python
#Extract rows 1 and 3.
print(arr[:3:2,:])
#Extract columns 2 and 4.
print(arr[:,1:4:2])
#Extract the elements: - 11 - 23 - 34 - 42
print("print 11 : ",arr[0,0])
print("print 23 : ",arr[1,2])
print("print 34 : ",arr[2,3])
print("print 42 : ",arr[3,1])

#Extract the four corner elements.
print("Corner Elements : ",arr[::3,::3])
#Reverse the row order using indexing.
print("Reverse the row : ",arr[::-1,:])
````

# 📖 2. Advanced Boolean Masking

## Dataset

Daily Temperatures (°C)

```
[32,35,38,29,27,
40,41,36,33,30,
28,31,39,42,37,
34,26,25,43,44]
```

---

## 🎯 Tasks

Without using loops:

- Find temperatures above 35°C.
- Find temperatures below 30°C.
- Find temperatures between 30°C and 38°C.
- Find indices where temperatures exceed 40°C.
- Count readings above 35°C.

---

````python
temperatures = np.array([32,35,38,29,27,
40,41,36,33,30,
28,31,39,42,37,
34,26,25,43,44])
print(temperatures)
````

````python
#Find temperatures above 35°C.
print("temperatures above 35°C : ",temperatures[temperatures>35])
#Find temperatures below 30°C.
print("temperatures below 30°C : ",temperatures[temperatures<30])
#Find temperatures between 30°C and 38°C.
print("temperatures between 30°C and 38°C : ",temperatures[(temperatures>30) & (temperatures<38)])
#Find indices where temperatures exceed 40°C.
print("indices where temperatures exceed 40°C : ",np.where(temperatures>40))
#Count readings above 35°C.
print("Count readings above 35°C : ",np.size(temperatures[temperatures>35]))
````

# 📖 3. Advanced Array Manipulation

## Dataset

Numbers from **1 to 36**

---

## 🎯 Tasks

Without using loops:

- Reshape into a 6×6 matrix.
- Extract the center 2×2 block.
- Extract only the border elements.
- Replace all even numbers with 0 (without modifying the original).
- Replace all values greater than 25 with -1.

---

````python
numbers = np.arange(1,37)
print(numbers)
````

````python
numbers = numbers.reshape((6,6))
#Reshape into a 6×6 matrix.
print("Reshape into a 6×6 matrix : ",numbers)
#Extract the center 2×2 block.
print("center 2×2 block : ",numbers[2:4,2:4])
#Extract only the border elements.
print("only the border elements : ",numbers[:6:5,:6:5])
#Replace all even numbers with 0 (without modifying the original).
copy = numbers.copy()
copy[copy%2==0]=0
print("all even numbers with 0 (without modifying the original) : ",copy)
#Replace all values greater than 25 with -1.
numbers[numbers>25]=-1
print("Replace all values greater than 25 with -1 : ",numbers)
````

### Broadcasting

Without modifying the original dataset:

---

````python
sales = np.array([
 [45,52,38,60,49],
 [48,55,40,63,51],
 [50,58,42,61,54],
 [47,54,39,65,53],
 [52,60,44,67,56],
 [49,57,41,64,55]
])
print(sales)
````

### Dataset Exploration

````python
#Find the shape.
print("shape : ",np.shape(sales))
#Number of weeks.
x,y = np.shape(sales)
print("Number of weeks : ",x)
#Number of products.
print("Number of products : ",y)
#Total data points.
print("Total data points : ",np.size(sales))b
````

### Sales Analysis

````python
#Total sales.
print("Total sales : ",np.sum(sales))
#Weekly sales totals.
print("Weekly sales totals : ",np.sum(sales,axis=1))
#Product-wise sales totals.
print("Product-wise sales totals : ",np.sum(sales,axis=0))
#Average weekly sales.
print("Weekly sales Average : ",np.mean(sales,axis=1))
#Average product sales.
print("Product sales Average : ",np.mean(sales,axis=0))
````

### Performance

````python
#Highest sale.
print("Highest sale : ",np.max(sales))
#Lowest sale.
print("Lowest sale : ",np.min(sales))
#Best-performing week.
print("Best-performing week : ",np.argmax(np.sum(sales,axis=1)))
#Best-performing product.
print("Best-performing product : ",np.argmax(np.sum(sales,axis=0)))
#Week with the lowest sales.
print("Week with the lowest sales : ",np.argmin(np.sum(sales,axis=1)))
````

### Boolean Masking

````python
#Sales greater than 60.
print("Sales greater than 60 : ",sales[sales>60])
#Products averaging above 50.
products_average = np.mean(sales,axis=0)
print("Products averaging above 50 : ",np.where(products_average>50))
#Weeks with total sales above the overall weekly average.
weekly_sales_total = np.sum(sales,axis=1)
overall_weekly_average = np.mean(np.sum(sales,axis=1))
print("overall_weekly_average : ",overall_weekly_average)
print("Weeks with total sales above the overall weekly average : ",np.where(weekly_sales_total>overall_weekly_average))
````

### Broadcasting

````python
#Add a ₹5 promotional boost to all sales.
pr = sales.copy().astype(float)
pr+=5
print(pr)
#Increase Product 3 sales by 10%.
pr[:,2] *= 1.10
print("Increase Product 3 sales by 10% : ",pr)
#Apply a 5% reduction to simulate returns.
pr*=0.95
print("5%\ reduction to simulate returns : ",pr)
````

# 🌟 Bonus Challenge – Cricket Tournament Analysis

## Scenario

Runs scored by 25 players during a tournament.

---

## Dataset

```
[78,45,102,66,89,
54,37,120,98,75,
81,64,59,110,43,
92,88,71,56,134,
47,69,83,95,61]
```

---

## 🎯 Tasks

### Statistics

- Mean
- Median
- Standard deviation
- Highest score
- Lowest score

---

### Performance Analysis

- Scores above 100.
- Scores below 50.
- Half-centuries (50–99).
- Centuries (100+).

---

### Ranking

- Top 5 scores.
- Bottom 5 scores.
- Highest scorer.
- Lowest scorer.

---

### Boolean Analysis

- Count scores above the average.
- Count even scores.
- Count odd scores.

---

````python
runs = np.array([78,45,102,66,89,
54,37,120,98,75,
81,64,59,110,43,
92,88,71,56,134,
47,69,83,95,61])
print(runs)
````

### Statistics

````python
#Mean
print("Mean = ",np.mean(runs))
#Median
print("Median = ",np.median(runs))
#Standard deviation
print("Standard deviation = ",np.std(runs))
#Highest score
print("Highest Score = ",np.max(runs))
#Lowest score
print("Lowest Score = ",np.min(runs))
````

### Performance Analysis

````python
#Scores above 100.
print("Scores above 100 : ",runs[runs>100])
#Scores below 50.
print("Scores below 50 : ",runs[runs<50])
#Half-centuries (50–99).
print("Half-centuries (50–99) : ", np.size(runs[(runs>=50)&(runs<100)]))
#Centuries (100+).
print("Centuries (100+) : ",np.size(runs[runs>=100]))
````

### Ranking

````python
#Top 5 scores.
print("Top 5 scores : ",np.sort(runs)[::-1][:5])
#Bottom 5 scores.
print("Bottom 5 scores : ",np.sort(runs)[:5])
#Highest scorer.
print("Highest Scorer : ", np.argmax(runs))
#Lowest scorer.
print("Lowest Scorer : ",np.argmin(runs))
````

### Boolean Analysis

````python
#Count scores above the average.
average_runs = np.mean(runs)
print("Count scores above the average : ", np.size(runs[runs>average_runs]))
#Count even scores.
print("Count even scores : ", np.size(runs[runs%2==0]))
#Count odd scores.
print("Count odd scores : ", np.size(runs[runs%2==1]))
````
