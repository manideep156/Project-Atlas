## 3. Boolean Masking

### What is Boolean Masking?

#### When is it useful?
#### Ans. We can use it for filtering without using loops
#### Why is it important in Data Analysis?
#### Ans. It makes complex operations and filtering easy

### Dataset

Numbers from 1 to 40

````python
arr= np.arange(1,41)
print(arr)
````

### Tasks

Return:

- Even numbers
- Odd numbers
- Numbers greater than 20
- Multiples of 3
- Multiples of both 2 and 5
- Numbers between 12 and 28
- Numbers not divisible by 4

````python
#Even numbers
mask_even = arr%2==0
print(arr[mask_even])
````

````python
#Odd numbers
mask_odd = arr%2==1
print(arr[mask_odd])
````

````python
#Numbers greater than 20
mask_20 = arr>20
print(arr[mask_20])
````

````python
mask_multiples_of_3 = arr%3==0
print(arr[mask_multiples_of_3])
````

````python
#Multiples of both 2 and 5
mask_multiples_of_2_and_5 = (arr%2==0) & (arr%5==0)
print(arr[mask_multiples_of_2_and_5])
````

````python
#Numbers between 12 and 28
mask_numbers_12_to_28 = (arr>11) & (arr<29)
print(arr[mask_numbers_12_to_28])
````

````python
#Numbers not divisible by 4
mask_not_by_4 = ~(arr%4==0)
print(arr[mask_not_by_4])
````

## 4. Random Number Generation

### Tasks

Generate:

- 10 random decimal numbers
- 4×4 random matrix
- 20 random integers
- Two identical arrays using the same seed

````python
#10 random decimal numbers
r=np.random.uniform(1,10,10)
print(r)
````

````python
#4×4 random matrix
r=np.random.uniform(1,10,(4,4))
print(r)
````

````python
#20 random integers
r=np.random.default_rng()
print(r.integers(1,101,20))
````

````python
#Two identical arrays using the same seed
# 1. Initialize the generator with a specific seed (e.g., 42)
rng1 = np.random.default_rng(seed=5)
array1 = rng1.integers(low=1, high=100, size=5)

# 2. Re-initialize a new generator with the EXACT same seed
rng2 = np.random.default_rng(seed=5)
array2 = rng2.integers(low=1, high=100, size=5)

print("Array 1:", array1)
print("Array 2:", array2)
````

# ⭐ Mini Challenge – Student Marks Analysis

## Scenario

You are given the marks of **8 students** in **3 subjects**.

- Each row represents a student.
- Each column represents a subject.

---

## Dataset

| Student | Subject 1 | Subject 2 | Subject 3 |
|---------|----------:|----------:|----------:|
| Student 1 | 78 | 82 | 91 |
| Student 2 | 65 | 70 | 68 |
| Student 3 | 90 | 88 | 95 |
| Student 4 | 55 | 60 | 58 |
| Student 5 | 84 | 79 | 81 |
| Student 6 | 73 | 69 | 77 |
| Student 7 | 92 | 94 | 90 |
| Student 8 | 48 | 52 | 57 |

Create this dataset as a NumPy array before solving the tasks.

---

# 🎯 Tasks

Without using loops:

### Basic Analysis

- Find the shape of the dataset.
- Find the number of students.
- Find the number of subjects.

---

### Statistics

- Find the overall average marks.
- Find the average marks of each student.
- Find the average marks of each subject.
- Find the highest mark.
- Find the lowest mark.

---

### Student Performance

- Find the total marks scored by each student.
- Identify the student with the highest total marks.
- Identify the student with the lowest total marks.

---

### Boolean Masking

Find:

- Students whose average marks are above 80.
- Students whose average marks are below 60.
- Students who scored above 90 in any subject.
- Students who scored above 75 in all subjects.

---

### Broadcasting

Without modifying the original dataset:

- Add 5 grace marks to every student.
- Add 2 bonus marks only to Subject 2.

---

````python
student_marks= np.array([
 [78,82,91],
 [65,70,68],
 [90,88,95],
 [55,60,58],
 [84,79,81],
 [73,69,77],
 [92,94,90],
 [48,52,57]
])
print(student_marks)
````

### Basic Analysis

````python
#shape of dataset
print(np.shape(student_marks))
````

````python
#number of students and subjects
students, subjects = np.shape(student_marks)
print("Number of students : ",students)
print("Number of subjects : ",subjects)
````

### Statistics

````python
# Find the overall average marks.
overall_mean = np.mean(student_marks)
print(overall_mean)
````

````python
# Find the average marks of each student.
student_average_marks = np.mean(student_marks,axis=1)
print(student_average_marks)
````

````python
# Find the average marks of each subject.
subject_average_marks = np.mean(student_marks,axis=0)
print(subject_average_marks)
````

````python
# Find the highest mark.
highest_mark = np.max(student_marks)
print(highest_mark)
````

````python
# Find the lowest mark.
lowest_mark = np.min(student_marks)
print(lowest_mark)
````

### Student Performance

````python
#Find the total marks scored by each student.
total_each_stu = np.sum(student_marks,axis=1)
print(total_each_stu)
````

````python
#Identify the student with the highest total marks.
stu_with_highest = np.argmax(total_each_stu)
print(stu_with_highest)
````

````python
#Identify the student with the lowest total marks.
stu_with_lowest = np.argmin(total_each_stu)
print(stu_with_lowest)
````

### Boolean Masking

````python
#Students whose average marks are below 60.
mask_1 = student_average_marks<60
print(np.where(mask_1)[0])
````

````python
#Students who scored above 90 in any subject.
mask_2 = np.max(student_marks,axis=1) > 90
print(np.where(mask_2)[0])
````

````python
#Students whose average marks are above 80.
mask_3 = student_average_marks>80
print(np.where(mask_3)[0])
````

````python
#Students who scored above 75 in all subjects.
mask_4 = (student_marks>75).all(axis=1)
print(np.where(mask_4)[0])
````

### Broadcasting

````python
#Add 5 grace marks to every student.
print(student_marks+5)
````

````python
#Add 2 bonus marks only to Subject 2.
student_marks[:, 1]+=2
print(student_marks)
````

Treat this as a one-dimensional NumPy array.

[28000, 35000, 42000, 50000, 38000,
 45000, 52000, 61000, 47000, 39000,

Create this dataset as a NumPy array before solving the tasks.

---

# 🎯 Tasks

Without using loops:

## Dataset Exploration

- Find the total number of employees.
- Find the highest salary.
- Find the lowest salary.
- Find the average salary.
- Find the median salary.
- Find the standard deviation.
- Find the variance.

---

## Salary Analysis

Find:

- Employees earning above ₹60,000.
- Employees earning below ₹40,000.
- Employees earning between ₹45,000 and ₹60,000 (inclusive).
- Employees earning exactly ₹50,000 (if any).

---

## Boolean Masking

Extract:

- Salaries greater than the average salary.
- Salaries less than the average salary.
- Salaries divisible by ₹1,000.
- Salaries divisible by ₹2,000.
- Salaries between ₹50,000 and ₹70,000.

---

## Vectorized Operations

Without modifying the original dataset:

- Give every employee a ₹2,500 bonus.
- Increase every salary by 8%.
- Deduct ₹1,000 from every salary.
- Calculate annual salary for every employee.
- Calculate monthly tax assuming a flat 10% deduction.

---

## Ranking

Find:

- Top 5 highest salaries.
- Bottom 5 lowest salaries.
- Difference between the highest and lowest salary.

---

````python
salaries = np.array([28000, 35000, 42000, 50000, 38000,
 45000, 52000, 61000, 47000, 39000,
 56000, 49000, 65000, 70000, 33000,
 41000, 53000, 62000, 58000, 46000,
 37000, 72000, 68000, 54000, 43000,
 59000, 76000, 48000, 51000, 64000])
print(salaries)
````

## Dataset Exploration

````python
#Find the total number of employees.
print(np.size(salaries))
````

````python
#Find the highest salary.
print(np.max(salaries))
````

````python
#Find the lowest salary.
print(np.min(salaries))
````

````python
#Find the average salary.
print(np.mean(salaries))
````

````python
#Find the median salary.
print(np.median(salaries))
````

````python
#Find the standard deviation.
print(np.std(salaries))
````

````python
#Find the variance.
print(np.var(salaries))
````

## Salary Analysis

````python
#Employees earning above ₹60,000.
mask_60k = salaries>60000
print(np.where(mask_60k)[0])
````

````python
#Employees earning below ₹40,000.
mask_below_40k = salaries<40000
print(np.where(mask_below_40k)[0])
````

````python
#Employees earning between ₹45,000 and ₹60,000 (inclusive).
mask_bw_45k_to_60k = (salaries>=45000) & (salaries<60000)
print(np.where(mask_bw_45k_to_60k)[0])
````

````python
#Employees earning exactly ₹50,000 (if any).
mask_50k = salaries==50000
print(np.where(mask_50k)[0])
````

## Boolean Masking

````python
#Salaries greater than the average salary.
mask_greater_avg = salaries>(np.average(salaries))
print(salaries[mask_greater_avg])
````

````python
#Salaries less than the average salary.
mask_lessthan_avg = salaries<(np.average(salaries))
print(salaries[mask_lessthan_avg])
````

````python
#Salaries divisible by ₹1,000.
mask_1000 = salaries%1000==0
print(salaries[mask_1000])
````

````python
#Salaries divisible by ₹2,000.
mask_2000 = salaries%2000==0
print(salaries[mask_2000])
````

````python
#Salaries between ₹50,000 and ₹70,000.
mask_sal_bw_50k_70k = (salaries>50000) & (salaries<70000)
print(salaries[mask_sal_bw_50k_70k])
````

## Vectorized Operations

````python
#Give every employee a ₹2,500 bonus.
print(salaries+2500)
````

````python
#Increase every salary by 8%.
salaries = salaries*1.08
print(salaries)
````

````python
#Deduct ₹1,000 from every salary.
salaries-=1000
print(salaries)
````

````python
#Calculate annual salary for every employee.
annual_sal = salaries*12
print(annual_sal)
````

````python
#Calculate monthly tax assuming a flat 10% deduction.
tax= salaries*0.1
print(tax)
````

## Ranking

````python
#Top 5 highest salaries.
print(np.sort(salaries)[-5:][::-1])
````

````python
#Bottom 5 lowest salaries.
print(np.sort(salaries)[:5])
````

````python
#Difference between the highest and lowest salary.
print(np.max(salaries)-np.min(salaries))
````
