## 🎯 NumPy Indexing & Slicing

```python
arr = np.arange(1, 26).reshape(5, 5)
print(arr)
```

### Practice

- First column
- Last column
- Middle element
- Top-left 2×2 block
- Bottom-right 2×2 block
- Rows 2–4
- Columns 2–4

```python
#First column
print(arr[:,0])
```

```python
#Last column
print(arr[:,-1])
```

```python
#Middle element
i,j=np.shape(arr)
print(arr[i//2 +1][j//2 +1])
```

```python
#Top-left 2×2 block
print(arr[:2,:2])
```

```python
#Bottom-right 2×2 block
print(arr[-2:,-2:])
```

```python
#Rows 2–4
print(arr[1:4])
```

```python
#Columns 2–4
print(arr[:,1:4])
```

# Day 2: NumPy Basics

## Reshaping Arrays

```python
arr=np.arange(1,21)
print(arr)
```

```python
#Reshape
reshaped_arr=arr.reshape(2,10)
print(reshaped_arr)
```

```python
#flatten
flattened_arr=reshaped_arr.flatten()
print(flattened_arr)
```

```python
#ravel
ravel_arr=reshaped_arr.ravel()
print(ravel_arr)
print(reshaped_arr)
```

```python
#Transpose
transpose=reshaped_arr.transpose()
print(transpose)
```

## 🎯 Aggregation Functions

```python
arr = np.arange(1, 13).reshape(3, 4)
print(arr)
```

```python
sum_arr=np.sum(arr)
print(sum_arr)
```

```python
mean_of_array=np.mean(arr)
print(mean_of_array)
```

```python
max_in_arr=np.max(arr)
print(max_in_arr)
```

```python
min_in_arr=np.min(arr)
print(min_in_arr)
```

```python
standard_deviation=np.std(arr)
print(standard_deviation)
```

```python
variance=np.var(arr)
print(variance)
```

```python
print(np.sum(arr, axis=0))
#adds the numbers in same column
```

```python
print(np.sum(arr, axis=1))
#adds the numbers in same row
```

```python
#calculates mean in same column
print(np.mean(arr, axis=0))
```

```python
#calculates mean in same row
print(np.mean(arr, axis=1))
```
