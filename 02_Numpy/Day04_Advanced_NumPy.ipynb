# 📖 1. Copy vs View

## Theory

Read about:

- View
- Copy
- Memory sharing
- `.base`
- `.copy()`

---

## Dataset

**Array A**

```
[10, 20, 30, 40, 50]
```

---

## 🎯 Tasks

Without using loops:

- Create a **view** of Array A.
- Create a **copy** of Array A.
- Modify the first element of the view.
- Observe the original array.
- Modify the second element of the copy.
- Observe the original array.
- Check whether the arrays share memory.

---

````python
import numpy as np
A = np.array([10, 20, 30, 40, 50])
print(A)
````

````python
#Create a **view** of Array A.
B=A
print(B)
````

````python
#Create a **copy** of Array A.
C=A.copy()
print(C)
````

````python
#Modify the first element of the view.
#Observe the original array.
B[0]=100
print(A)
````

````python
#Modify the second element of the copy.
#Observe the original array.
C[1]=100
print("Original array : ",A)
print("Copy :",C)
````

# 📖 2. Sorting Arrays

## Theory

Read about:

- `sort()`
- `argsort()`

---

## Dataset

```
[45, 12, 87, 23, 56, 91, 34, 67, 18, 72]
```

---

## 🎯 Tasks

Without using loops:

- Sort the array in ascending order.
- Sort the array in descending order.
- Find the indices that would sort the array.
- Find the three largest values.
- Find the three smallest values.

---

````python
dataset =  np.array([45, 12, 87, 23, 56, 91, 34, 67, 18, 72])
print("Unsorted Dataset : ", dataset)
````

````python
#sort in ascending
sorted_dataset = np.sort(dataset)
print("Ascendingly Sorted Dataset : ", sorted_dataset)
````

````python
#sort in descending
desc_sorted_dataset = np.sort(dataset)[::-1]
print(desc_sorted_dataset)
````

````python
#Find indices that would sort the array
print(np.argsort(dataset))
````

````python
#Find three largest values
print(desc_sorted_dataset[:3])
````

````python
#Find three smallest Values
print(sorted_dataset[:3])
````

# 📖 3. Unique Values

## Theory

Read about:

- `unique()`

---

## Dataset

```
[2, 4, 6, 2, 8, 6, 4, 10, 12, 8, 6, 14, 16, 2, 18]
```

---

## 🎯 Tasks

Without using loops:

- Find all unique values.
- Count the number of unique values.
- Count the frequency of each value.
- Identify the most frequent value.

---

````python
unique_dataset = np.array([2, 4, 6, 2, 8, 6, 4, 10, 12, 8, 6, 14, 16, 2, 18])
print(unique_dataset)
````

````python
#Find all unique values
unique_values = (np.unique(unique_dataset))
print(unique_values)
````

````python
#Count the number of unique values
print(np.size(unique_values))
````

````python
#Count the frequency of each value
unique_values, counts = (np.unique(unique_dataset, return_counts=True))
print(counts)
````

````python
#Find the most frequent value
index_of_most_frequent_unique_value = np.argmax(counts)
print(unique_values[index_of_most_frequent_unique_value])
````

# 📖 4. Searching Arrays

## Theory

Read about:

- `where()`
- `argmax()`
- `argmin()`

---

## Dataset

```
[12, 45, 67, 23, 89, 45, 90, 34, 12, 78, 65, 89]
```

---

## 🎯 Tasks

Without using loops:

- Find all indices where the value is **45**.
- Find indices of values greater than **70**.
- Find the position of the maximum value.
- Find the position of the minimum value.
- Find the indices of all even numbers.

---

````python
array = np.array([12, 45, 67, 23, 89, 45, 90, 34, 12, 78, 65, 89])
print(array)
````

````python
#Find all indices where the value is 45
print(np.where(array==45))
````

````python
#Find all indices greater than 75
print(np.where(array>70))
````

````python
#Find position of max value
print(np.argmax(array))
````

````python
#Find position of min value
print(np.argmin(array))
````

````python
#Find indices of all even numbers
print(np.where(array%2==0))
````

# ⭐ Mini Challenge – Sales Data Analysis

## Scenario

A retail store recorded product sales for 20 consecutive days.

---

## Dataset

```
[120, 135, 150, 120, 180,
 175, 160, 150, 145, 170,
 200, 195, 180, 175, 160,
 155, 145, 140, 135, 130]
```

---

## 🎯 Tasks

### Analysis

- Find total sales.
- Find average daily sales.
- Find highest sales.
- Find lowest sales.
- Find median sales.
- Find standard deviation.

---

### Searching

- Find all days where sales were **180**.
- Find days where sales were above the average.
- Find days where sales were below **150**.

---

### Unique Values

- Find unique sales values.
- Count the frequency of each value.
- Identify the most frequent sales value.

---

### Sorting

- Display the top 5 highest sales values.
- Display the bottom 5 lowest sales values.

---

````python
sales_data = np.array([120, 135, 150, 120, 180,
 175, 160, 150, 145, 170,
 200, 195, 180, 175, 160,
 155, 145, 140, 135, 130])
print(sales_data)
````

### Analysis

````python
#Find total sales.
print("total sales = ", np.sum(sales_data))
#Find average daily sales.
print("Average Daily Sale = ",np.mean(sales_data))
#Find highest sales.
print("Highest Sales = ",np.max(sales_data))
#Find lowest sales.
print("Lowest Sales = ",np.min(sales_data))
#Find median sales.
print("Median Sales = ",np.median(sales_data))
#Find standard deviation.
print("Standard Deviation = ",np.std(sales_data))
````

### Searching

````python
#Find all days where sales were **180**.
days_where_sales_were_180 = np.where(sales_data==180)
print("days where sales were 180 : ",days_where_sales_were_180)
#Find days where sales were above the average.
abv_avg = np.where(sales_data>np.mean(sales_data))
print("days where sales were above the average : ",abv_avg)
#Find days where sales were below **150**.
below_150 = np.where(sales_data<150)
print("days where sales were below 150 : ",below_150)
````

### Unique Values

````python
#Find unique sales values.
print("Unique sales Value : ", np.unique(sales_data))
#Count the frequency of each value.
print("Frequecy of each element : ",np.unique_counts(sales_data))
````

### Sorting

````python
#Display the top 5 highest sales values.
print("top 5 sales values are : ",np.sort(sales_data)[::-1][:5])
#Display the bottom 5 lowest sales values.
print("bottom 5 sales values are : ",np.sort(sales_data)[:5])
````

# 🌟 Bonus Challenge – Movie Ratings Analysis

## Scenario

A movie platform collected ratings from 30 users.

---

## Dataset

```
[4.5, 3.8, 4.9, 4.2, 3.5,
 5.0, 4.8, 4.1, 3.9, 4.7,
 4.3, 3.6, 4.4, 4.0, 4.6,
 4.8, 3.7, 4.9, 4.2, 3.8,
 4.1, 5.0, 4.7, 4.3, 3.9,
 4.6, 4.4, 3.5, 4.8, 4.2]
```

---

## 🎯 Tasks

### Statistics

- Find the average rating.
- Find the highest rating.
- Find the lowest rating.
- Find the median rating.
- Find the standard deviation.

---

### Filtering

Find ratings:

- Greater than **4.5**
- Less than **4.0**
- Between **4.0** and **4.8** (inclusive)

---

### Searching

- Find all indices where the rating is **5.0**
- Find all indices where the rating is **4.8**

---

### Unique Values

- Find all unique ratings.
- Count the frequency of each rating.
- Identify the most common rating.

---

### Sorting

- Sort ratings in ascending order.
- Sort ratings in descending order.
- Display the top 5 ratings.

---

### Statistics

````python
movie_ratings = np.array([4.5, 3.8, 4.9, 4.2, 3.5,
 5.0, 4.8, 4.1, 3.9, 4.7,
 4.3, 3.6, 4.4, 4.0, 4.6,
 4.8, 3.7, 4.9, 4.2, 3.8,
 4.1, 5.0, 4.7, 4.3, 3.9,
 4.6, 4.4, 3.5, 4.8, 4.2])
print(movie_ratings)
````

````python
#Find the average rating.
print("Average Rating : ",np.mean(movie_ratings))
#Find the highest rating.
print("Highest Rating : ",np.max(movie_ratings))
#Find the lowest rating.
print("Lowest Rating : ",np.min(movie_ratings))
#Find the median rating.
print("Median Rating : ",np.median(movie_ratings))
#Find the standard deviation.
print("Standard Deviation : ",np.std(movie_ratings))
````

### Filtering

````python
#Greater than **4.5**
print("Greater than 4.5 : ", movie_ratings[movie_ratings>4.5])
#Less than **4.0**
print("Less than 4.0 : ",movie_ratings[movie_ratings<4.0])
#Between **4.0** and **4.8** (inclusive)
print("Between 4.0 and 4.8 (inclusive) : ", movie_ratings[(movie_ratings>=4.0)&(movie_ratings<=4.8)])
````

### Searching

````python
#Find all indices where the rating is 5.0
print("all indices where the rating is 5.0 : ", np.where(movie_ratings==5.0))
#Find all indices where the rating is 4.8
print("all indices where the rating is 4.8 : ", np.where(movie_ratings==4.8))
````

### Unique Values

````python
#Find all unique ratings.
print("all unique ratings : ", np.unique(movie_ratings))
#Count the frequency of each rating.
unique_ratings, counts = np.unique_counts(movie_ratings)
print("frequency of each rating. : ", counts)
#Identify the most common rating.
print("the most common rating : ", unique_ratings[np.argmax(counts)])
````

### Sorting

````python
#Sort ratings in ascending order.
print("ratings in ascending order : ", np.sort(movie_ratings))
#Sort ratings in descending order.
print("ratings in descending order : ",np.sort(movie_ratings)[::-1])
#Display the top 5 ratings.
print("top 5 ratings : ", np.sort(movie_ratings)[::-1][:5])
````
