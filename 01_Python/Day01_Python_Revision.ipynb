## *args and **kwargs

```python
def sum_all(*args):
    return sum(args)
```

```python
sum_all(1, 2, 3)
sum_all(10, 20, 30, 40)
```

```python
def student_details(**kwargs):
    print(kwargs)
```

```python
student_details(name="Manideep", age=24, city="Vizag")
```

```python
def student_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)
student_details(name="Manideep", age=24, city="Vizag")
```

## Lambda Functions

Small anonymous functions used for short operations.

```python
#Exercise 1: Square numbers
nums = [1, 2, 3, 4, 5]
print(list(map(lambda x: x*x , nums)))
```

```python
#Exercise 2: Even filter; Filter even numbers from:
nums = [10, 15, 20, 25, 30, 35]
print(list(filter(lambda x: x%2==0, nums)))
```

```python
#Exercise 3: Sorting; Sort by second value:
data = [(1, 5), (3, 2), (9, 1), (4, 7)]
print(sorted(data,key=lambda x:x[1]))
```

## List Comprehensions

A compact way to create lists using loops in a single line.

```python
#Exercise 1: Squares
nums = [1, 2, 3, 4, 5]
squares = [x * x for x in nums]
```

```python
#Exercise 2: Even numbers

#From 1 to 50, get all even numbers.

a= [x for x in range(1,51) if x%2==0]
print(a)
```

```python
#Exercise 3: Word lengths
words = ["data", "science", "python", "ai"]

#Get lengths of each word.
print(list(map(len,words)))
```

```python
#Exercise 4: Filter + transform

#From 1 to 20:

#keep only even numbers
#square them

a=[x*x for x in range(1,21) if x%2==0]
print(a)
```
