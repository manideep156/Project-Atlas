## 💡 Biggest Challenges

### 1. Simulating a 12% Increase

One of the more challenging parts was applying a **12% sales increase only to products whose average sales were below the overall average**.

```python
total_average = np.mean(sales)

prods_below_overall = np.where(product_average < total_average, 1.12, 1.0)

new_sales = sales * prods_below_overall
```

### How it works

* `np.mean(sales)` calculates the overall average of the dataset.
* `product_average` stores the average sales for each product.
* `np.where()` creates a multiplier array:

  * `1.12` for products below the overall average.
  * `1.00` for all other products.
* NumPy broadcasting automatically applies these multipliers to each product column without using loops.

**Concepts Used**

* Aggregation
* Boolean masking
* Broadcasting
* Vectorized operations

---

### 2. Counting Product 4 Sales Above 95

```python
prod4_95 = (sales[:,3] > 95)

np.sum(prod4_95)
```

### How it works

* `sales[:,3]` selects all rows from Product 4.
* The comparison `> 95` creates a Boolean array.
* `True` values represent weeks where Product 4 sold more than 95.
* `np.sum()` counts the number of `True` values because NumPy treats `True` as `1`.

**Concepts Used**

* Column selection
* Boolean masking
* Aggregation

---

### 3. Identifying Weeks Where Every Product Sold Above 70

```python
np.where(np.all(sales > 70, axis=1))[0]
```

### How it works

* `sales > 70` creates a Boolean matrix.
* `np.all(..., axis=1)` checks every row.
* A row becomes `True` only if **every product** satisfies the condition.
* `np.where()` returns the indices of those weeks.

**Concepts Used**

* Boolean masking
* Logical aggregation
* Row-wise analysis
